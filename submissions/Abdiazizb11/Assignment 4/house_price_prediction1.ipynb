```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_csv("clean_house_l5_dataset.csv")
df.head(5)
```

```python
x=df.drop(columns=["Price","LogPrice"])
y=df["Price"]
```

```python
#spliting data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#predicting by using both LR AND RF
lr=LinearRegression()
rf=RandomForestRegressor(n_estimators=100, random_state=42)
lr.fit(x_train,y_train)
rf.fit(x_train,y_train)
lr_pred=lr.predict(x_test)
rf_pred=rf.predict(x_test)
```

```python
#model comparring
def model(name,y_true,y_predect):
    r2=r2_score(y_true,y_predect)
    mae=mean_absolute_error(y_true,y_predect)
    mse=mean_squared_error(y_true,y_predect)
    rmse=np.sqrt(mse)
    print(name)
    print(f"r2: {r2:,.2f}")
    print(f"mae: {mae:,.0f}")
    print(f"mse: {mse:,.0f}")
    print(f"rmse: {rmse:,.0f}")

model("LinearRegression",y_test,lr_pred)
model("RandomForest",y_test,rf_pred)
    
```

```python
# Select the index
i = 4
sample_x = x_test.iloc[[i]]
actual_price = y_test.iloc[i]

lr_val = lr.predict(sample_x)[0]
rf_val = rf.predict(sample_x)[0]

print(f"--- Sanity Check (Row {i}) ---")
print(f"Actual Price:  ${actual_price:,.2f}")
print(f"LR Prediction: ${lr_val:,.2f}")
print(f"RF Prediction: ${rf_val:,.2f}")
```
